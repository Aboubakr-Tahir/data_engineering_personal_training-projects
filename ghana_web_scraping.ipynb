ill try now to extract the hyper link

```python
#/html/body/div/section[2]/div/div/div/div/div[2]/div/div[1]/div[1]/a 
#html > body > div > section:nth-of-type(2) > div > div > div > div > div:nth-of-type(2) > div > div:nth-of-type(1) > div:nth-of-type(1) > a
extracted_element = soup.select_one('html > body > div > section:nth-of-type(2) > div > div > div > div > div:nth-of-type(2) > div > div:nth-of-type(1)')
link_element = extracted_element.find('a',href=True) 
hyper_element = link_element['href']
hyper_element
```

we extracted just the first one for now , lets extract all of the boxes in this page and put it in a pandas dataframe to looke beautiful

```python
#/html/body/div/section[2]/div/div/div/div/div[2]/div 
extracted_boxes = soup.select_one('html > body > div > section:nth-of-type(2) > div > div > div > div > div:nth-of-type(2) > div')
list_boxes = extracted_boxes.find_all('div',class_='list-wrap') 
df_list = []
for box in list_boxes : 
    iterative_list = []
    title = box.find('div',class_='list-title') 
    iterative_list.append(title.text)
    link = box.find('a',href=True)
    iterative_list.append(link['href']) 
    description = box.find('div',class_='list-desc') 
    iterative_list.append(description.text) 
    date = box.find('div',class_='list-date')
    iterative_list.append(date.text.strip()) 
    agency = box.find('div',class_='list-agency') 
    iterative_list.append(agency.text.strip())  
    df_list.append(iterative_list) 
    
import pandas as pd 
columns = ['title','link','description','date','agency']
df = pd.DataFrame(df_list , columns=columns)
df   
    
```

now for each hyperlink we extract , we need to access and scrap its information one by one

this is the code to extract the information from the hyperlinks 

```python
#/html/body/div/section[2]/div/div/div/div/div[2]/div --- hl = hyperlink
url = requests.get('https://tenders.ppa.gov.gh/tenders/31729')
soup_of_hl = BeautifulSoup(url.text , 'html.parser') 
extracted_hl_info = soup_of_hl.select_one('html > body > div > section:nth-of-type(2) > div > div > div > div > div:nth-of-type(2) > div')
list_infos=extracted_hl_info.find_all('dl')
lista=[]
for info in list_infos :
    lista.append(info.text.strip())  
   
       
```

now we need to add it into the a code that goes through all of the hyper links one by one

```python
extracted_boxes = soup.select_one('html > body > div > section:nth-of-type(2) > div > div > div > div > div:nth-of-type(2) > div')
list_boxes = extracted_boxes.find_all('div',class_='list-wrap') 
df_list = []
for box in list_boxes : 
    iterative_list = []
    title = box.find('div',class_='list-title') 
    iterative_list.append(title.text)
    link = box.find('a',href=True)
    #here is the part where we access the hyper link infos
    hyper_link=link['href'] 
    if hyper_link :
        url = requests.get(hyper_link)
        soup_of_hl = BeautifulSoup(url.text , 'html.parser') 
        extracted_hl_info = soup_of_hl.select_one('html > body > div > section:nth-of-type(2) > div > div > div > div > div:nth-of-type(2) > div')
        list_infos=extracted_hl_info.find_all('dl')
        lista=[]
        for info in list_infos :
            lista.append(info.text.strip())  
    iterative_list.append(hyper_link)  
    for information in lista :
        iterative_list.append(information) 
    description = box.find('div',class_='list-desc') 
    iterative_list.append(description.text) 
    date = box.find('div',class_='list-date')
    iterative_list.append(date.text.strip()) 
    agency = box.find('div',class_='list-agency') 
    iterative_list.append(agency.text.strip())  
    df_list.append(iterative_list) 
    
    
df_list 
```
