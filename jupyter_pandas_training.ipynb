Merge , Join , and Concatenate 

```python
df1=pd.read_csv(r"c:\Users\hp\Downloads\LOTR.csv")
df1
```

```python
df2=pd.read_csv(r"c:\Users\hp\Downloads\LOTR 2.csv")
df2 
```

```python
df1.merge(df2 , how='inner', on=['FellowshipID','FirstName'])  
df1.merge(df2)
```

```python
df1.merge(df2 , how='outer') 
```

```python
df1.merge(df2 , how='left')
```

```python
df1.merge(df2,how='right' , on='FellowshipID')
```

```python
df1.merge(df2 , how='cross')
```

```python
df1.set_index('FellowshipID').join(df2.set_index('FellowshipID'),lsuffix='_left',rsuffix='_right',how='left')
```

```python
pd.concat([df1.set_index('FellowshipID'),df2.set_index('FellowshipID')] , join='outer' , axis=1) 
```
